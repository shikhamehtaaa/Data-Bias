```python
#create a dictionary pairing comments with whether or not they were marked as toxic
toxic = df["toxic"]
print(toxic)
toxic_list = toxic.tolist()
dictionary = {}
for key in comments_list:
    for value in toxic_list:
        dictionary[key] = value
        toxic_list.remove(value)
        break
        
#print some of the dictionary to ensure that the lists were paired correctly
dict(list(dictionary.items())[0:5])
```

```python
#find the most frequent word in the comments marked as toxic

#create a list of the comments marked as toxic
toxic_comments = []
for key, value in dictionary.items():
    if value == "yes":
        toxic_comments.append(key)
print(toxic_comments[0:5]) 

#convert the list into a long string and make all of the characters lowercase
toxic_comments_string = (" ").join(toxic_comments)
toxic_comments_string = toxic_comments_string.lower()

#remove stop words from the string
STOPWORDS = set(stopwords.words('english'))
for x in toxic_comments_string:
    if x not in STOPWORDS:
        toxic_comments_list.append(x)
print(toxic_comments_list[0:5])

#create a dictionary that pairs words with their frequencies
toxic_comments_dict = {}
for x in toxic_comments_list:
    if x not in toxic_comments_dict.keys():
        toxic_comments_dict[x] = 1
    else:
        toxic_comments_dict[x] += 1
sorted_toxic_comments_dict = sorted(toxic_comments_dict.items(), key=lambda x:x[1], reverse=True)
print(sorted_toxic_comments_dict)
```

```python
#find the most frequent word in the comments not marked as toxic

#create a list of the comments not marked as toxic
nontoxic_comments = []
for key, value in dictionary.items():
    if value == "no":
        nontoxic_comments.append(key)
print(nontoxic_comments[0:5]) 

#make sure that there are only strings in the list before joining all of the elements
strings_only = []
for i in nontoxic_comments:
    if type(i) is str: 
        strings_only.append(i)
print(strings_only[0:5])

#convert the list into a long string and make all of the characters lowercase
nontoxic_comments_s = (" ").join(strings_only)
nontoxic_comments_s = nontoxic_comments_s.lower()
nontoxic_comments_string = list(nontoxic_comments_s.split(" "))
print(nontoxic_comments_string[0:5])

#remove stop words from the string
STOPWORDS = set(stopwords.words('english'))
for x in nontoxic_comments_string:
    if x not in STOPWORDS:
        nontoxic_comments_list.append(x)
print(nontoxic_comments_list[0:5])

#create a dictionary that pairs words with their frequencies
nontoxic_comments_dict = {}
for x in nontoxic_comments_list:
    if x not in nontoxic_comments_dict.keys():
        nontoxic_comments_dict[x] = 1
    else:
        nontoxic_comments_dict[x] += 1
sorted_nontoxic_comments_dict = sorted(nontoxic_comments_dict.items(), key=lambda x:x[1], reverse=True)
print(sorted_nontoxic_comments_dict)
```
